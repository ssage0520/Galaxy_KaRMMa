```python
import sys

sys.path.append("../")
import h5py as h5

from karmma import KarmmaConfig
from karmma.utils import plot_1pt_linear, plot_1pt_log, plot_corr, plot_pseudo_cl
```

```python
configfile = "/spiff/ssage/karmma_data/LN_dg_theta/Inputs/desy3.yaml"
config = KarmmaConfig(configfile)

analysis = config.analysis
io = config.io
mcmc = config.mcmc

nside = analysis.nside
nbins = analysis.nbins
io_dir = io.io_dir
n_samples = mcmc.n_samples

with h5.File(f"{io_dir}/summary_statistics.h5", "r") as f:
    # bins
    corr_bin_centres = f["bins/corr_bin_centres"][:]
    corr_bin_edges = f["bins/corr_bin_edges"][:]
    eff_ell = f["bins/eff_ell"][:]
    ell_edges = f["bins/ell_edges"][:]
    linear_bins = f["bins/linear_bins"][:]
    log_bins = f["bins/log_bins"][:]

    # truth
    corr_true = f["truth/corr"][:]
    pseudo_cl_true = f["truth/pseudo_cl"][:]
    pdf_linear_true = f["truth/pdf_linear"][:]
    pdf_log_true = f["truth/pdf_log"][:]

    # samples
    corr_samples = f["samples/corr"][:]
    corr_errors_samples = f["samples/corr_errors"][:]
    pseudo_cl_samples = f["samples/pseudo_cl"][:]
    pdf_linear_samples = f["samples/pdf_linear"][:]
    pdf_log_samples = f["samples/pdf_log"][:]
```

```python
plot_corr(corr_samples, corr_true, corr_bin_centres, interval=95)
```

```python
plot_pseudo_cl(pseudo_cl_samples, pseudo_cl_true, eff_ell, nside, interval=95)
```

```python
plot_1pt_linear(pdf_linear_samples, pdf_linear_true, linear_bins, interval=95)
```

```python
plot_1pt_log(pdf_log_samples, pdf_log_true, log_bins, interval=95)
```
